#### 3.2.1.3 세트

세트는 수학의 집합을 구현한 자료형.  
세트에 저장된 값들은 순서가 없기 때문에 인덱싱과 슬라이싱 불가.  
대신 중복 값 저장이 불가능하다는 특징을 활용해 중복 제거용으로 사용하거나  
교집합, 합집합, 차집합과 같은 집합 개념이 필요한 경우 사용

>사용법
* {값1, 값2}
* set(반복가능객체)
<br>

다만 빈 세트 사용 시에는 중괄호를 사용할 수 없음. (이는 딕셔너리)
빈 세트 생성 시 set 함수 이용

````python
# 중복 제거 
a = [1,2,2,4,4,5]
a = set(a)
print(a)
a = list(a)
print(a)
````

````python
#  

a = {1,2,3}
b = {3,4,5}

print(a.difference(b))
print(a.difference(a))
print(a.union(b))
print(a.intersection(b))

print('-----------------')
print(a - b)
print(b - a)
print(a & b)
print(a | b)
````

````python
# jaccard similarity
string1 = '나는 파이썬을 좋아합니다'
string2 = '나는 파이썬을 정말 좋아합니다'

splitted_string1 = string1.split()
splitted_string2 = string2.split()

print(splitted_string1)
print(splitted_string2)

splitted_string1_set = set(splitted_string1)
splitted_string2_Set = set(splitted_string2)

print(splitted_string1_set & splitted_string2_Set)
print(splitted_string1_set | splitted_string2_Set)

print(len(splitted_string1_set & splitted_string2_Set) / len(splitted_string1_set | splitted_string2_Set))
````

**※주의**

들여쓰기 규칙은 다음과 같음  
1. 공백이나 탭을 이용하여 들여쓰기 수행

````python
# 점수는 0~100 90점을 넘어가면 a
# 점수는 0~100 80점을 넘어가면 b
# 점수는 0~100 70점을 넘어가면 c
# 점수는 0~100 60점을 넘어가면 d
# 점수는 0~100 60점을 미만으로나온다면 f

def get_grade(score):
    if score > 100 or score< 0:
        return '유효하지않는 점수입니다.'
    if  score  >= 90:
        return "a"
    elif score >= 80:
        return "b"
    elif score >= 70:
        return "c"
    elif score >= 60:
        return "d"
    else:
        return "f"
````

**가변 매개변수**
함수로 전달해야 하는 인수의 개수가 정해지지 않아도  
매개변수 선언 가능한데 이를 가변 매개변수라고 함  

가변 매개변수를 만드는 키워드는 애스터리스크(*)이며 개수에 따른 차이가 있음

1. def show(*args):  
&nbsp;&nbsp;&nbsp;&nbsp;for item in args:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;print(item)  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;show('happy', 'birthday')  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-> 인자를 튜플로 전달

2. def show(**kwargs):  
&nbsp;&nbsp;&nbsp;&nbsp;for key, item in args:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;print(key, item)

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;show(a=1, b=2)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-> 인자를 딕셔너리로 전달

````python
# 점수는 0~100 90점을 넘어가면 a
# 점수는 0~100 80점을 넘어가면 b
# 점수는 0~100 70점을 넘어가면 c
# 점수는 0~100 60점을 넘어가면 d
# 점수는 0~100 60점을 미만으로나온다면 f

def get_grade(score):
    if score > 100 or score< 0:
        return '유효하지않는 점수입니다.'
    if  score  >= 90:
        return "a"
    elif score >= 80:
        return "b"
    elif score >= 70:
        return "c"
    elif score >= 60:
        return "d"
    else:
        return "f"
````

````python
def show(*args): # 에스터 리스크:무조건 튜플로 받아준다.
    print(args)
    for item in args:
        print(item)
````

````python
def show(a,b,*args): # 에스터 리스크:무조건 튜플로 받아준다.
    print(args)
    print(a)
    print(b)
    for item in args:
        print(item)
````

````python
show(60,70,80,90,100)
````

````python
def show(**kwargs): # 값은 딕셔너리로 만들어진다 별은 한개더 사용한다.
    print(kwargs)
    for key,value in kwargs.items():
        print(key,value)
````

````python
show(a=10, b=20, c=30, d=40, e=50)
````
