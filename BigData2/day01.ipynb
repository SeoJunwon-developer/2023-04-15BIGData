````python
# rank : 1~100등까지 등수가 담긴 array
# 1 ~ 20: 1
# 21 ~ 40: 2
# 41 ~ 60 : 3
# 61 ~ 80 : 4
# 81 ~ 100 : 5
# 의 값을 부여
````

````python
ranks = np.random.choice(np.arange(1,101), size=100)
````

````python
ranks
````

````python
# solution
grades = []
for rank in ranks:
    if rank > 80:
        grades.append(5)
    elif rank > 60:
        grades.append(4)
    elif rank > 40:
        grades.append(3)
    elif rank > 20:
        grades.append(2)
    else:
        grades.append(1) 
grades
````

````python
# solution2
((ranks - 1) // 20) + 1
````

````python
# 함수 만들기
# 함수의 입력은 복권 구매 금액
# 금액에 맞는 복권 매수 구매
# 복권은 자동임
# 1등번호는 6,11,16,19,21,32
# 복권 번호는 1~ 45
win_number = [6,11,16,19,21,32]
````

````python
def buy_lottery_tickets(amount):


    num_tickets = amount // 1000
    lottery_numbers = np.random.choice(range(1, 46), size=num_tickets*6, replace=True).reshape(num_tickets, 6)
    amount = 5000 
    lottery_numbers = buy_lottery_tickets(amount)
    print("구매한 복권 매수:", lottery_numbers.shape[0])
    print("당첨 번호:", win_number)
    print("복권 번호:", lottery_numbers)
````

````python
win_number = [6,11,16,19,21,32]
def lotto(money):
    count = 0 
    for _ in range(money // 1000):
        my_numbers = list(np.random.randint(1,46, 6))
        my_numbers.sort()
        my_numbers
        if win_number == my_numbers:
            count += 1
    return count
````

````python
lotto(100000000000)
````
